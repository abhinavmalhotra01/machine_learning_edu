```python
# eqn ->  y = Bo + B1X1 + B2X1^2 +....
# its called linear ->  coeff combination is linear in y
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:,1:2].values  # we need only level col
Y = dataset.iloc[:,2:3].values
print(Y)
```

```python
plt.scatter(X,Y)
```

```python
# it is visible -> values need exponents 
# no training , testing set needed 
# feature scaling taken care by library

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,Y)
```

```python
# for polynomial
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2) # a transformer tool that wil transform X into a new matrix (X_poly) containing X1,X1^2,X1^3....
# degree -> 2 -> X1^2 created
X_poly = poly_reg.fit_transform(X)
print(X)
X_poly 
poly_reg.fit(X_poly,Y)
 # first col -> X1^0 , second col -> X1^1 , third col -> X1^2 
```

```python
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,Y)
```

```python
plt.scatter(X,Y,color="green")
plt.plot(X,lin_reg.predict(X),color="red") # predictions of linear reg model
plt.title("Linear Regression")
plt.show()
```

```python
# from above we can see that lin_reg model is not a good model to predict in this case
# so using polynomial reg
plt.scatter(X,Y,color="green")
plt.plot(X,lin_reg_2.predict(poly_reg.fit_transform(X)))
plt.title("polynomial regression")
```

```python
# curve -> because polynomial is non-linear -> cost function is way lesser than linear reg model
# now we wanna do even bettern -> degree 3
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=3) # a transformer tool that wil transform X into a new matrix (X_poly) containing X1,X1^2,X1^3....
X_poly = poly_reg.fit_transform(X)
# print(X)
# X_poly
poly_reg.fit(X_poly,Y)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,Y)
plt.scatter(X,Y,color="green")
plt.plot(X,lin_reg_2.predict(poly_reg.fit_transform(X)))
plt.title("polynomial regression")
```

```python
# now the model is even better 
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4) # a transformer tool that wil transform X into a new matrix (X_poly) containing X1,X1^2,X1^3....
X_poly = poly_reg.fit_transform(X)
# print(X)
# X_poly
poly_reg.fit(X_poly,Y)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,Y)
plt.scatter(X,Y,color="green")
plt.plot(X,lin_reg_2.predict(poly_reg.fit_transform(X)))
plt.title("polynomial regression")
```

```python
# now it is almost passing through all the points
# for more adv
X_grid = np.arange(min(X),max(X),0.1)
X_grid=X_grid.reshape((len(X_grid),1))
plt.scatter(X,Y,color="green")
plt.plot(X_grid,lin_reg_2.predict(poly_reg.fit_transform(X_grid)))
plt.title("polynomial regression")
```

```python
# more continous curve - we got

print(lin_reg.predict([[6.5]]))
print(lin_reg_2.predict(poly_reg.fit_transform([[6.5]])))
```
