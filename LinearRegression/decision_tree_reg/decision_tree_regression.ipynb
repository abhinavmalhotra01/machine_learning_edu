```python
# each node -> feature ; each branch -> decision ; each leaf -> outcome (categorcl/continous)
# algo to build -> CART(Gini index) ; ID3(IG,entropy) 
# H(S)=sum(-p(c)*log(p(c))) ; S-> current set ; C -> set of classes in S (eg:yes,no) ; p(c) -> ratio of number of elements in c to number of elem in S ;;; IG(A)=sum(p(c)*H(c)) -> avg of all entropies over the children c of feature A
# choosing which split/feature -> highest IG (entropy) -> IG(A,S)=H(S)-sum(p(t)*H(t))  -> split S into feature A ; H(S)-> entropy of S ; T -> subsets created from splitting S by feature A ; p(t)-> ratio of no of elements in t to elements in S ; H(t)-> entropy of subset t
# choose the A with largest IG
# for regression -> lets say we have a split of X>20 -> 2 choices for each number -> yes or no -> goes into same as classificn 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt                       
```

```python
dataset=pd.read_csv("Position_Salaries.csv")
X=dataset.iloc[:,1:2].values
Y=dataset.iloc[:,2:3].values
# feature scaling by lib
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X,Y)
```

```python
print(regressor.predict([[6.5]]))
```

```python
plt.scatter(X,Y,color="red")
plt.plot(X,regressor.predict(X))
plt.show()
```

```python
# above is a trap -> regression template is problem -> this model is not continous -> in above case , it simply joined the Y points 
# inc resolution :
X_grid = np.arange(min(X),max(X),0.01)
X_grid=X_grid.reshape((len(X_grid),1))
plt.scatter(X,Y,color="green")
plt.plot(X_grid,regressor.predict(X_grid))
# plt.title("polynomial regression")
```
