```python
# simple linear regression 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:,:-1].values
Y = dataset.iloc[:,1].values
print(X)
print(Y)
```

```python
# missing,categorical are not there in given dataset

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=1/3,random_state=0)
# print(X_train)

#feature scaling will be taken care by regression library
```

```python
# from sklearn we will import linear_model library
# in this library - import LinearRegression class
# create object of LinearRegression  class
# this obj is regressor - which we will fit to training set
# fit - method/tool 

from sklearn.linear_model import LinearRegression
regressor = LinearRegression(); #obj of linear regression class
regressor.fit(X_train,Y_train); # fits regressor obj to training set

# regressor - machine and we made this machine learn on the dataset of X_train,Y_train

```

```python
# predicting the Test set results
Y_pred = regressor.predict(X_test);
# regressor has learnt from training sets , now we are using it to predict on x_test

# Y_test is having the real values - in this case - it is the set having the salaries that those 10 employees were actually given\
# Y_pred is set having the predicted salaries from our regressor obj
# obv values should be close/not much far
# sometimes very far observations may also be predicted

print(Y_pred)
```

```python
# now visualization :)
# x-axis - years of exp , y-axis - salary

# scatter - scatter obsv points
plt.scatter(X_train,Y_train,color='red') # (x,y) args in scatter

# plotting regression - plt plot - we r not scattering points
plt.plot(X_train,regressor.predict(X_train),color='green')
# plotted regression line bw x_train and predicted values by regressor obj
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()  # just final that we r ready to see 
```

```python
# now plotting test set results

# scatter - real values 
plt.scatter(X_test,Y_test,color='red');
plt.plot(X_test,Y_pred,color='green');  # even if we didnt change this - plot same - line toh ek hi h - chaahe test set se bnao ya train set se
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
