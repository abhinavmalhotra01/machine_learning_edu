```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Mall_Customers.csv")

x = dataset.iloc[:,[3,4]].values
dataset.head()
```

```python
# dendogram :
import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(x,method='ward'))
plt.title("dendogram")
plt.xlabel("customers")
plt.ylabel("euclidean distances")
plt.show()
```

```python
# 3 clusters ->  good 
# fitting it ->
from sklearn.cluster import AgglomerativeClustering  # bottom-up
hc_3 = AgglomerativeClustering(n_clusters=3,affinity="euclidean",linkage="ward")
```

```python
y_3 = hc_3.fit_predict(x)
```

```python
# graph - works only for 2D
plt.scatter(x[y_3==0,0],x[y_3==0,1],s=50,c='red',label='cluster 1')
plt.scatter(x[y_3==1,0],x[y_3==1,1],s=50,c='green',label='cluster 2')
plt.scatter(x[y_3==2,0],x[y_3==2,1],s=50,c='blue',label='cluster 3')
plt.show()
```
