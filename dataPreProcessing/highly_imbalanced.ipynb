```python
# we will do Under-Sampling on the dataset
# highly imbalanced meaning :
            # lets say , there is categorical data - example : spam and not spam
            # now what if in the data set - spam mails are in lakhs whereas not spam mails are in hundereds
            # in such a case , the data set is said to be imbalanced 
            # it is imp to take care of such a data , since it may create problems in training models and so on
```

```python
import numpy as np
import pandas as pd

credit_card_data = pd.read_csv('credit_data.csv')
```

```python
# legit and fraud credit card  - classification problem
credit_card_data.head()

# class - 0,1
# 0 - legit
# 1 - fraud 
```

```python
# distribution of 2 classes :
credit_card_data['Class'].value_counts() # values in col with name 'Class'
```

```python
# as it is evident - 2 lakh legit and 492 fraud - highly imbalanced

# separating the legit and fraudulent transaction
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]

print(legit.shape)
print(fraud.shape)
```

```python
# under-sampling 

# building a sample dataset containing similar distribution of legit and fraudulent transactions
# number of fraud trans -> 492

legit_sample = legit.sample(n=492)  # will take 492 sample values from legit df and stored them in legit_sample df

print(legit_sample.shape)
```

```python
# concatenate the 2 dataframes
new_dataset = pd.concat([legit_sample,fraud],axis=0)  # joining legit and fraud; axis =0 -> vertical concat -> first 492 legit after that 492 fraud
new_dataset.head() 
```
