```python
import openai
import urllib.request
from PIL import Image
import cv2
# import cv2_imshow
```

```python
openai.api_key = "hehe"
```

```python
def gen_img(img_input):
    
    img_response = openai.Image.create(
        prompt= img_input,
        n=1,
        size="256x256"
    )

    img_url = img_response['data'][0]['url']

    urllib.request.urlretrieve(img_url,'img.png')

    img=Image.open('img.png')

    return img
```

```python
img_input=input("enter details of img")
out_img =gen_img(img_input)

from IPython.display import Image
Image(filename='img.png') 
```
