```python


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Social_Network_Ads.csv")
```

```python
X=data.iloc[:,:-1].values
Y=data.iloc[:,2].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
#transforming to higher dimension that somehow the data becomes linearly separable
# however this mapping to higher dimensn can be highly computer intensive 

# the kernel trick - gaussian rbf kernel 
# in same dimension using k(x,l)=e^||x-l||^2/2*(sigma)^2
# just adjusting decision boundary -> still being in the lower dimension -> we check k(x,l) -> put constraints(>,<,=) on it -> to divide points in differ classes
# k(x,l1)+k(x,l2) pr bhi constraints daal skte ; can extend even further ig

# types of kernel functions -> 1. gaussian rbf (upr wala)
#                              2. sigmoid kernel (logistic wala)
#                              3. polynomial kenrel (y.Xy + r)^d -> kindof polynomial graphs depending on decided 'd'
#                              other also exist :/

from sklearn.svm import SVC
classifier = SVC(kernel='rbf',random_state=0)  # gaussian rbf (default bhi yhi h) 
classifier.fit(X_train,Y_train)  
```

```python
Y_pred=classifier.predict(X_test)
print(Y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,Y_pred)
print(cm)


import seaborn as sns
plt.figure(figsize=(5,5))
sns.heatmap(cm,annot=True)
plt.xlabel('pred')
plt.ylabel('truth')
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_pred))
```

```python
# visualizing training set results
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), Y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.50),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.50))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('SVC (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
# on training set
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_test), Y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.5),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.5))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('SVC (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
